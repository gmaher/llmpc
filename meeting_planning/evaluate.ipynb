```python
pivoted = df.pivot(index='model', columns='people', values='accuracy')
pivoted = pivoted.reindex(['GPT-4o', "LLMPC T=5 K=1", "LLMPC T=10 K=1", "LLMPC T=15 K=1", "LLMPC T=3 K=3", "LLMPC T=9 K=3"])
```

```python
pivoted.T.plot(kind='bar', figsize=(8, 4))

plt.title('Meeting Planning Success Rate')
plt.xlabel('Number of meetings')
plt.ylabel('Score')
plt.legend(title='Model')
plt.tight_layout()
plt.savefig("./output/meeting_planning_bar.pdf", dpi=300)
plt.show()
```
