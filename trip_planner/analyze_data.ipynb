```python
size_groups = {}
for k,v in data.items():
    v['key'] = k
    num_cities = len(v['cities'].split("**"))
    v['num_cities'] = num_cities
    if num_cities not in size_groups:
        size_groups[num_cities] = []
    size_groups[num_cities].append(v)
```

```python
samples = 30
selected = []
for k,l in size_groups.items():
    s = np.random.choice(l, size=samples, replace=False).tolist()
    selected+=s
```

```python
len(selected)
```

```python
new_data = {}
for d in selected:
    k = d['key']
    new_data[k] = d
```

```python
with open("./data/trip_planning_reduced.json",'w') as f:
    json.dump(new_data,f,indent=1)
```
