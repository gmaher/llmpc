```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
mpc = pd.read_csv("./output/mpc.csv")
llmpc = pd.read_csv("./output/llmpc.csv")
```

```python
dt = 0.1
t = np.arange(llmpc.shape[0])*dt
```

```python
plt.figure(figsize=(12,3))
plt.plot(t,mpc['x'], label='x - MPC', color='b')
plt.plot(t,mpc['v'], label='v', color='g')
plt.plot(t,llmpc['x'], label='x - LLMPC', color='b', linestyle='--')
plt.plot(t,llmpc['v'], label='v', color='g', linestyle='--')
plt.xlabel("t")
plt.ylabel('position, velocity')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('./output/state.pdf',dpi=300)
plt.show()
```

```python
plt.figure(figsize=(12,3))
plt.plot(t,mpc['u'], label='MPC', color='k')
plt.plot(t,llmpc['u'], label='LLMPC', color='k', linestyle='--')
plt.ylabel('u')
plt.xlabel('t')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("./output/control.pdf",dpi=300)
plt.show()
```

```python
plt.figure(figsize=(12,3))
plt.plot(t,mpc['c'], label='MPC', color='r')
plt.plot(t,llmpc['c'], label='LLMPC', color='r', linestyle='--')
plt.ylabel('cost')
plt.xlabel('t')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("./output/cost.pdf",dpi=300)
plt.show()
```
