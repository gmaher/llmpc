```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("./output/planner_comparison_K1_H3.csv")
df2 = pd.read_csv("./output/planner_comparison_K5_H3.csv")
df3 = pd.read_csv("./output/planner_comparison_K10_H3.csv")
df4 = pd.read_csv("./output/planner_comparison_K15_H3.csv")
```

```python
plt.figure(figsize=(12,3))
plt.hist(df1.llm_costs, bins=20, alpha=0.75, label='LLMPC K=1')
plt.hist(df2.llm_costs, bins=20, alpha=0.75, label='LLMPC K=5')
plt.hist(df3.llm_costs, bins=20, alpha=0.75, label='LLMPC K=10')
plt.hist(df4.llm_costs, bins=20, alpha=0.75, label='LLMPC K=15')
plt.hist(df1.mpc_costs, bins=20, alpha=0.75, label='MPC K=1')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,3))
plt.hist(df1.llm_costs/df1.mpc_costs, bins=20, alpha=0.75, label='LLMPC K=1')
plt.hist(df2.llm_costs/df2.mpc_costs, bins=20, alpha=0.75, label='LLMPC K=5')
plt.hist(df3.llm_costs/df3.mpc_costs, bins=20, alpha=0.75, label='LLMPC K=10')
plt.hist(df4.llm_costs/df4.mpc_costs, bins=20, alpha=0.75, label='LLMPC K=15')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(9,3))
plt.hist(df1.llm_costs/df1.mpc_costs, bins=20, alpha=0.75, label='LLMPC K=1')
plt.hist(df2.llm_costs/df2.mpc_costs, bins=20, alpha=0.75, label='LLMPC K=5')
plt.hist(df3.llm_costs/df3.mpc_costs, bins=20, alpha=0.75, label='LLMPC K=10')
plt.hist(df4.llm_costs/df4.mpc_costs, bins=20, alpha=0.75, label='LLMPC K=15')
plt.legend()
plt.ylabel('Frequency')
plt.xlim(0,10)
plt.xlabel("LLPC cost/MPC cost")
plt.tight_layout()
plt.savefig("./output/spring_compare_bar.pdf", dpi=300)
plt.show()
```
